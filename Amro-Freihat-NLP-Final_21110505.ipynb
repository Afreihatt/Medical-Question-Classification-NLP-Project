```python

def is_arabic(text):
    if not isinstance(text, str) or not text.strip():
        return False
    try:
        return detect(text) == 'ar'
    except:
        return False


arabic_df = df_sample[df_sample['question_body'].apply(is_arabic)]
print(f"Original dataset size: {len(df_sample)}")
print(f"After filtering for Arabic only: {len(arabic_df)}")
```

```python
def preprocess_arabic_text(text):

    if not isinstance(text, str):
        return []



    text = re.sub(r'[إأآا]', 'ا', text)  # Normalize Alef
    text = re.sub(r'[يى]', 'ي', text)    # Normalize Yaa
    text = re.sub(r'ة', 'ه', text)       # Normalize Taa Marbouta
    text = re.sub(r'[ؤئء]', '', text)    # Remove Hamza""


    # Remove diacritics (harakat)
    text = re.sub(r'[\u064B-\u0652]', '', text)



    text = re.sub(r'[^\w\s]', ' ', text)


    text = re.sub(r'\d+', '', text)


    # Apply tokenization - using simple split instead of word_tokenize to avoid dependency issues
    words = text.split()



    arabic_stopwords = set(stopwords.words('arabic'))
    words = [word for word in words if word not in arabic_stopwords and len(word) > 1]



    stemmer = SnowballStemmer("arabic")
    words = [stemmer.stem(word) for word in words]

    return words
```

```python
from sklearn.metrics import classification_report, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dropout, Dense
import numpy as np
all_results_list=[]
def build_and_evaluate_model(model_type, X_train, y_train, X_test, y_test, embedding_name):

    X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))


    model = Sequential([
        Input(shape=(1, X_train.shape[1])),
        model_type(128, return_sequences=True),
        model_type(64),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    experiment_name = f"{embedding_name} + {model_type.__name__}"
    print(f"\n--- RUNNING EXPERIMENT: {experiment_name} ---")

    model.fit(X_train_reshaped, y_train, epochs=10, batch_size=128, verbose=0)


    y_pred = np.argmax(model.predict(X_test_reshaped), axis=1)


    report_dict = classification_report(y_test, y_pred,
                                        target_names=label_encoder.classes_.astype(str),
                                        zero_division=0,
                                        output_dict=True)
    all_results_list.append({'Experiment': experiment_name, 'Accuracy': report_dict['accuracy']})

    return report_dict
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

results_list = []


models_to_test = [SimpleRNN, LSTM]


for embedding_name, (X_train, X_test) in embeddings_data.items():
    for model_type in models_to_test:


        report = build_and_evaluate_model(model_type, X_train, y_train, X_test, y_test, embedding_name)


        experiment_name = f"{embedding_name} + {model_type.__name__}"
        results_list.append({
            'Experiment': experiment_name,
            'Accuracy': report['accuracy'],
            'F1-Score (Macro Avg)': report['macro avg']['f1-score']
        })


print("\n\n--- All Experiments Complete. Final Results: ---")
results_df = pd.DataFrame(results_list)
```
